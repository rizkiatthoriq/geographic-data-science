```python
# Function to find 3 nearest neighbors for each labeled data point
def find_knn_for_labels(df_labeled, df_unlabeled, k=3):
    required_columns = ['2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023', '2024', 'Euclidean Distance'] #+'EuclideanDistance'

    # Ensure required columns are present in both DataFrames
    for col in required_columns:
        if col not in df_labeled.columns:
            raise KeyError(f"Column '{col}' not found in labeled data.")
        if col not in df_unlabeled.columns:
            raise KeyError(f"Column '{col}' not found in unlabeled data.")

    results = []

    knn = NearestNeighbors(n_neighbors=k)
    knn.fit(df_unlabeled[required_columns])

    for index, row in df_labeled.iterrows():
        label = row['PP ID']
        data_point = row[required_columns].values.reshape(1, -1)

        # Debug statement to check the data point
        print(f"Processing label {label} with data point: {data_point}")

        distances, indices = knn.kneighbors(data_point)
        nearest_neighbors = df_unlabeled.iloc[indices[0]].reset_index(drop=True)

        # Add label to the neighbors DataFrame
        nearest_neighbors['PP ID'] = label
        nearest_neighbors['distance'] = distances[0]

        results.append(nearest_neighbors)

    return pd.concat(results).reset_index(drop=True)
```

```python
# prompt: I want the list of KKN Result for EVI_2015 to EVI_2024 if then calculated
# SDM (Standard Difference Mean), the value between 0.20 and 0.245. SDM Equation as follows:
# Absolute (Average EVI in year t in pp data -  Average EVI in year t in cp data)/Square Root ((Squared of Standard Deviation of EVI in year t in pp data + Squared of Standard Deviation of EVI in year t in pp data)/2).

import os
from google.colab import drive
import pandas as pd
from sklearn.neighbors import NearestNeighbors

# drive.mount('/content/drive')

# os.chdir('/content/drive/My Drive/KNN')
# Import the datasets
pp = pd.read_csv('project_plot_si_values.csv',delimiter=';')
cp = pd.read_csv('control_plot_si_values.csv',delimiter=';')
pp['EuclideanDistance'] = pp['EuclideanDistance']/1000
cp.columns
# cp.rename(columns={"Euclidean Distance (Km)": "EuclideanDistance"}, inplace=True)
# cp = cp[cp['EuclideanDistance']<=5]
# Celan NaN in cp
cp.dropna(inplace=True)

# Function to find 3 nearest neighbors for each labeled data point
def find_knn_for_labels(df_labeled, df_unlabeled, k=3):
    required_columns = ['EVI_2015', 'EVI_2020', 'EVI_2024', 'EuclideanDistance']

    # Ensure required columns are present in both DataFrames
    for col in required_columns:
        if col not in df_labeled.columns:
            raise KeyError(f"Column '{col}' not found in labeled data.")
        if col not in df_unlabeled.columns:
            raise KeyError(f"Column '{col}' not found in unlabeled data.")

    results = []

    knn = NearestNeighbors(n_neighbors=k)
    knn.fit(df_unlabeled[required_columns])

    for index, row in df_labeled.iterrows():
        label = row['PP_ID']
        data_point = row[required_columns].values.reshape(1, -1)

        distances, indices = knn.kneighbors(data_point)
        nearest_neighbors = df_unlabeled.iloc[indices[0]].reset_index(drop=True)

        # Add label to the neighbors DataFrame
        nearest_neighbors['PP_ID'] = label
        nearest_neighbors['distance'] = distances[0]

        results.append(nearest_neighbors)

    return pd.concat(results).reset_index(drop=True)

# Finding 3 nearest neighbors for each labeled data point
knn_results = find_knn_for_labels(pp, cp, k=1)

# Calculate SDM for each year and filter results
sdm_filtered_results = []
for year in ['EVI_2015', 'EVI_2020', 'EVI_2024']:
    pp_mean = pp[year].mean()
    cp_mean = knn_results[year].mean()
    pp_std = pp[year].std()
    cp_std = knn_results[year].std()
    sdm = abs(pp_mean - cp_mean) / ((pp_std**2 + cp_std**2) / 2)**0.5
    if 0.20 <= sdm <= 0.245:
        sdm_filtered_results.append((year, sdm, knn_results))

# Display filtered results
for year, sdm, result in sdm_filtered_results:
    print(f"Year: {year}, SDM: {sdm:.3f}")
    display(result)
```
